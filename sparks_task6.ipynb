**Importing Necessary Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets as datasets
from sklearn import tree
import warnings
warnings.filterwarnings("ignore")
```

**Loading Data**

```python
iris=datasets.load_iris()
```

converting data into dataframe

```python
df=pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df
```

**Target variable**

```python
iris.target
```

**Basic information**

```python
df.info()
```

```python
df.describe()
```

**Cheking Null values**

```python
df.isna().sum()
```

No Null values

**Scatter Plot: Sepal Length vs Sepal Width**

```python
df.plot.scatter(x='sepal length (cm)', y='sepal width (cm)')
plt.title('Scatter Plot: Sepal Length vs Sepal Width')
plt.show()
```

**Pair Plot**

```python
pd.plotting.scatter_matrix(df.drop('petal width (cm)', axis=1), c=df['petal width (cm)'])
plt.suptitle('Pair Plot')
plt.show()
```

**Histograms**

```python
df.hist()
plt.suptitle('Histograms')
plt.show()
```

**Box Plot**

```python
df.plot.box()
plt.title('Box Plot')
plt.show()
```

**Violin Plots: Petal Length and Petal Width by Species**

```python
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
sns.violinplot(x=iris.target, y='petal length (cm)', data=df)
plt.subplot(1, 2, 2)
sns.violinplot(x=iris.target, y='petal width (cm)', data=df)
plt.suptitle('Violin Plots: Petal Length and Petal Width by Species')
plt.show()
```

**spliting data into train and test**

```python
x=df.iloc[:,:]
y=iris.target
```

**Model Building**

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x,y)
```

using decision tree classifier to form a tree

**Tree Map**

```python
fig = plt.figure(figsize=(10, 8))
_ = tree.plot_tree(dt, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
plt.show()
```
